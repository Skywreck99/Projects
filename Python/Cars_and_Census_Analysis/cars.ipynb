# Task 3 - Worksheet 1
##### Purpose - Creates a code that calculates the average mpg city driving and the average mpg for highway driving for each manufacturer
##### Output - Creates a list of two lists. One list for the city and the other one will be for the highway


```python
biglist = list()
city_ave = list()
hw_ave = list()

for m in manu:
    summation = 0
    num_of_m = 0
    
    for key in mpg:
        if key['manufacturer'] == m:
            summation += float(key['cyl'])
            num_of_m += 1
    city_ave.append((summation/num_of_m))

for mn in manu:
    summation += 0
    num_of_mn = 0
    
    for key in mpg:
        if key['manufacturer'] == mn:
            summation += float(key['hwy'])
            num_of_mn += 1
    hw_ave.append((summation/num_of_mn))

biglist.append(city_ave)
biglist.append(hw_ave)

biglist
        
```
